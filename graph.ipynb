```python
import numpy as np
import matplotlib.pyplot as plt
 
# set width of bar
barWidth = 0.25
fig = plt.subplots(figsize =(12, 8))
 
# set height of bar

he_multi= [5.830466985702515,4.036669969558716,3.1565401554107666,2.8970491886138916]
he_joblib= [6.07314920425415,3.67317795753479,3.167772054672241,2.725414991378784]
 
# Set position of bar on X axis
br1 = np.arange(len(he_multi))
br2 = [x + barWidth for x in br1]
br3 = [x + barWidth for x in br2]
 
# Make the plot

plt.bar(br2,he_multi, color ='g', width = barWidth,
        edgecolor ='grey', label ='he_multi')
plt.bar(br3,he_joblib, color ='b', width = barWidth,
        edgecolor ='grey', label ='he_joblib')
 
# Adding Xticks
plt.xlabel('cores', fontweight ='bold', fontsize = 15)
plt.ylabel('Time', fontweight ='bold', fontsize = 15)
plt.title("he -multiprocessing and joblib")
plt.xticks([r + barWidth for r in range(len(he_multi))],
        ['2', '4', '6', '8'])
 
plt.legend()
plt.show()
```
